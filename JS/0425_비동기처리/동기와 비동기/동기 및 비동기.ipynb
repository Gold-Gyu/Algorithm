# 동기식 : 하나의 작업을 시작하면 그 작업이 끝나기 전에는 다음 작업 안함. 앞 작업이 끝나야 다음 작업 가능

1. 순차대로만 작업
2. 오래걸리는 작업이 있다면 계속 기다려야하는 단점이 있다.

# 비동기식 : 동시에 여러일을 하는 것처럼 보이지만, 사실 시간을 쪼개서 일을 한다. 일단 작업을 시작한다.

- 요청과 응답을 동기식으로 처리할 수 없음.
1. 작업을 시작한 후 결과를 기다리지 않고 다음 작업을 처리하는 것
2. 시간이 필요한 작업들은 요청을 보낸 뒤 응답이 빨리 오는 작업부터 처리
3. 사용자 경험 측면에서 매우 중요한 것이 비동기식
4. 즉 동기식 처리는 특정 로직이 실행되는 동안 다른 로직 실행을 차단하기 때문에 마치 프로그램이 응답하지 않는 듯한 사용자 경험을 만들게 됨
5. 비동기로 처리한다면 먼저 처리되는 부분부터 보여줄 수 있다 ==> 사용자 경험에 긍정적 효과
